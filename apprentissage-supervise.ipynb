On va, pour le reste des caractéristiques, utiliser la **matrice de corrélation** pour verifier la corrélation tout en ayant d'abord supprimer les valeurs aberrantes

```python
new_df = df[['price', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'zipcode', 'sqft_living15', 'sqft_lot15']]
```

```python
new_df.shape
```

```python
#on supprime maintenant les valeurs aberrantes pour pour éviter des erreurs sur la matrice de corrélation
# fonction pour le calcul des différents grandeur du boxplot
def get_iqr_values(df_in, col_name): 
    median = df_in[col_name].median()
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    
    iqr = q3 - q1
    minimum = q1 - 1.5*iqr
    maximum = q3 + 1.5*iqr
    return median, q1, q3, iqr, minimum, maximum
#fonction pour supprimer les outliers
def remove_outliers(df_in, col_name):
    _, _, _, _, minimum, maximum = get_iqr_values(df_in, col_name)
    df_out = df_in.loc[(df_in[col_name] > minimum) & (df_in[col_name] < maximum)]
    return df_out
# récupérons les colonnes dans new_df
colonnes = new_df.columns.tolist()
# aplliquons nos fonction à chacunes de nos colonnes dans new_df
for elt in colonnes:
    df_out = remove_outliers(new_df, elt)
df_out.shape
```

```python
#Determinons maitenant la matrice de corrélation
import seaborn as sns
def plot_correlation_map( df ):
    corr = df.corr()
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
        )
plot_correlation_map(df_out)
```

### Interprétation
le coefficient de corrélation est une valeur compris entre -1 et 1
- une valeur proche de 1 représent un corrélation positive
- une valeur proche de -1 représente une corrélation négative
- une valeur proche de 0 représente une corrélation nulle

Sur cette matrice, on retrouve des valeurs proche de 1 et d'autre qui lui sont éloignés
on va donc prendre les corrélation avec des coefficient suprieure à 0.5 en valeur absolue

Divisons notre model en ensemble de training et test

Ici on a choisit pour le reste du traite de ne pas supprimer les valeurs aberrantes

```python
#Pour la regression lineaire on choisirait 'sqft_living' comme variable prédictif
from sklearn.model_selection import train_test_split
X = df['sqft_living'].values[:, np.newaxis]
Y = df['price'].values
#séparation en train et test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 40)
```

```python
#Pour la regression mutltilineaire
Xm = df[['sqft_living', 'bathrooms', 'lat', 'grade', 'sqft_above', 'sqft_living15']]
Ym = df['price'].values
#on normalise maintenant nos variable prédictives
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
Xm_scaled = scale.fit_transform(Xm[['sqft_living', 'bathrooms', 'lat', 'grade', 'sqft_above', 'sqft_living15']])
```

```python
#appliquons la regression lineaire à notre ensemble d'entrainement
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
model = LinearRegression()
model.fit(X_train, Y_train)

predicted = model.predict(X_test)
print("MSE", mean_squared_error(Y_test, predicted))
print("R Squared", metrics.r2_score(Y_test, predicted))
```

```python
#traçons maintenant la droite de rgression
plt.scatter(X, Y, color ="r")
plt.title("linear Regression")
plt.xlabel("Surface Habitable")
plt.ylabel("Prix")
plt.plot(X, model.predict(X), color="k")
plt.show()
```

```python
#Appliquons une regressions multilineaire
Xm_train, Xm_test, Ym_train, Ym_test = train_test_split(Xm, Ym,  test_size = 0.35, random_state = 40)
model = LinearRegression()
model.fit(Xm_train, Ym_train)
predicted = model.predict(Xm_test)

print("MSE", mean_squared_error(Ym_test, predicted))
print("R Squared", metrics.r2_score(Ym_test, predicted))
```

### comparaison entre regression lineaire et multilineaire
le R_squared de la regression lineaire et de la regression mutilineaire qui sont égaux a peu près à 0.5 indique que la moitié de leurs sortie peut être expliquée par leurs entrée.
le MSE de la regression multilineaire inferieur à celui de la regression lineraire impliquue que la regression lineaire minimise le plus les erreurs

```python
#la regression polynomial
from sklearn.preprocessing import PolynomialFeatures
Xm_train, Xm_test, Ym_train, Ym_test = train_test_split(Xm, Ym,  test_size = 0.35, random_state = 40)
lg = LinearRegression()
poly = PolynomialFeatures(degree=3)
Xm_train_fit = poly.fit_transform(Xm_train)
lg.fit(Xm_train_fit, Ym_train)
Xm_test_ = poly.fit_transform(Xm_test)
predicted = lg.predict(Xm_test_)
print("MSE", mean_squared_error(Ym_test, predicted))
print("R Squared", metrics.r2_score(Ym_test, predicted))
```

### Comparaison entre les différents types de regression
